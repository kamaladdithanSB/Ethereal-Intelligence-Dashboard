```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


url = "https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv"

try:
    df = pd.read_csv(url)
    df.columns = ['CustomerID', 'Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']

    print("🚀 Success! Data loaded perfectly.")
    print(df.head())

    X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    print("✅ Features scaled and ready for K-Means.")

except Exception as e:
    print(f"Still hitting a wall: {e}")
    print("Plan B: Upload the file manually to Colab using the folder icon on the left!")
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)


df['Cluster'] = y_kmeans


cluster_map = {
    0: 'Standard (Middle Class)',
    1: 'Ideal Targets (High Income/High Spend)',
    2: 'Sensible (Low Income/Low Spend)',
    3: 'Careless (Low Income/High Spend)',
    4: 'Elites (High Income/Low Spend)'
}
df['Segment'] = df['Cluster'].map(cluster_map)
```

```python
import plotly.graph_objects as go


cluster_profile = df.groupby('Segment')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean().reset_index()

fig = go.Figure()


for i, segment in enumerate(cluster_profile['Segment']):
    fig.add_trace(go.Scatterpolar(
        r=cluster_profile.iloc[i, 1:].values,
        theta=['Age', 'Income', 'Spending Score'],
        fill='toself',
        name=segment
    ))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 100])),
    showlegend=True,
    template='plotly_dark',
    title="Customer Segment Personalities (The Logic)"
)

fig.show()
```
